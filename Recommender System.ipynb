Now we have to convert this format 

'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'

into this format - 

['Action','Adventure','Fantasy','Science Fiction']


```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
# #its a module that will convert string into list
# import ast
# ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'].apply(convert)
```

converting string into list for genres

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head(2)
```

converting string into list for cast

```python
#function
def convert2(obj):
    L2=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L2.append(i['name'])
            counter+=1
        else:
            break
    return L2
```

```python
movies['cast'].apply(convert2)
# movies.head(2)
```

```python
movie_up=movies
movie_up.head(2)
```

```python
movie_up['cast']=movie_up['cast'].apply(convert2)
movie_up.head(2)
```

converting string into list for crew

```python
def fetchdir(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```
